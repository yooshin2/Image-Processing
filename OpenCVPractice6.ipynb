```
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = np.full((512, 512, 3), 255, np.uint8)
image = cv2.line(image, (0, 0), (255, 255), (0, 0, 255), 5)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = np.full((512, 512, 3), 255, np.uint8)
image = cv2.rectangle(image, (100, 100), (255, 255), (255, 0, 0), 5)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = np.full((512, 512, 3), 255, np.uint8)
image = cv2.circle(image, (255, 255), 50, (255, 0, 255), -1)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = np.full((512, 512, 3), 255, np.uint8)
image = cv2.ellipse(image, (255, 255), (100, 60), 0, 0, 360, (255, 0, 0), 3)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = np.full((512, 512, 3), 255, np.uint8)
points1 = np.array([[10, 10], [100, 300], [300, 400], [400, 400]])
points2 = np.array([[300, 40], [60, 100], [150, 300], [260, 400], [300, 240]])
image = cv2.polylines(image, [points1], True, (0, 0, 255), 4)
image = cv2.polylines(image, [points2], True, (255, 0, 255), 10)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = np.full((512, 512, 3), 255, np.uint8)
image = cv2.putText(image, 'Hello World', (100, 150), cv2.FONT_ITALIC, 1.5, (0, 0, 255))
image = cv2.putText(image, 'Hello World', (100, 350), cv2.FONT_ITALIC, 1.5, (0, 0, 255), 8)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
