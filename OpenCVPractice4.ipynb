```
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('image3.jpg', cv2.IMREAD_GRAYSCALE)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_GRAY2RGB))
plt.show()

images = []

ret, thr1 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
ret, thr2 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY_INV)
ret, thr3 = cv2.threshold(image, 127, 255, cv2.THRESH_TRUNC)
ret, thr4 = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO)
ret, thr5 = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO_INV)

images.append(thr1)
images.append(thr2)
images.append(thr3)
images.append(thr4)
images.append(thr5)

for i in images:
    plt.imshow(cv2.cvtColor(i, cv2.COLOR_GRAY2RGB))
    plt.show()

```

```
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('handwriting.jpg', cv2.IMREAD_GRAYSCALE)

res, thr1 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
thr2 = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 15, 2) 
thr3 = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 15, 2)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_GRAY2RGB))
plt.show()

plt.imshow(cv2.cvtColor(thr1, cv2.COLOR_GRAY2RGB))
plt.show()

plt.imshow(cv2.cvtColor(thr2, cv2.COLOR_GRAY2RGB))
plt.show()

plt.imshow(cv2.cvtColor(thr3, cv2.COLOR_GRAY2RGB))
plt.show()

```
