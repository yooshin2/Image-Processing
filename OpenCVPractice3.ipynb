```
import cv2
import matplotlib.pyplot as plt

image1 = cv2.imread('img_1.jpg')
image2 = cv2.imread('img_2.jpg')

print(image1.shape)
print(image2.shape)

image1 = cv2.resize(image1, (559, 314), interpolation=cv2.INTER_CUBIC)

plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
plt.show()

plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
plt.show()

result1 = cv2.add(image1, image2) # cv2.add() : Saturation 연산 0보다 작으면 0, 255보다 크면 255로 표현
plt.imshow(cv2.cvtColor(result1, cv2.COLOR_BGR2RGB))
plt.show()

result2 = image1 + image2 # np.add() : Modulo 연산 256은 0, 257은 1, 258은 2
plt.imshow(cv2.cvtColor(result2, cv2.COLOR_BGR2RGB))
plt.show()
```
