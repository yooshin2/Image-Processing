```
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('people1.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

result1 = cv2.resize(image, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_CUBIC)
plt.imshow(cv2.cvtColor(result1, cv2.COLOR_RGB2BGR))
plt.show()

result2 = cv2.resize(image, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_AREA)
plt.imshow(cv2.cvtColor(result2, cv2.COLOR_BGR2RGB))
plt.show()
```

```
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('people1.jpg')

height, width = image.shape[:2]

M = np.float32([[1, 0, 100], [0, 1, 10]])
result = cv2.warpAffine(image, M, (width+100, height+100))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
```

```
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('people1.jpg')

height, width = image.shape[:2]

M = cv2.getRotationMatrix2D((width / 2, height / 2), 90, 0.5)
result = cv2.warpAffine(image, M, (width, height))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
```
